## Step 6.3 – Convert Credit Limit to Numeric

The `credit_limit` column contains currency symbols (`$`). Before performing numerical analysis, the currency symbol is removed and the column is converted into a numeric data type.

```python
if df["credit_limit"].dtype == "object":
    df["credit_limit"] = (
        df["credit_limit"]
        .str.replace("$", "", regex=False)
        .astype(float)
    )

print(df["credit_limit"].dtype)
```

## Step 13.1 – Credit Limit Category

### Purpose

Customers are categorized into Low, Medium, and High credit limit groups based on their available credit limit.

```python
df["credit_limit_category"] = pd.cut(
    df["credit_limit"],
    bins=[0,5000,15000,float("inf")],
    labels=["Low","Medium","High"]
)

df[["credit_limit","credit_limit_category"]].head()
```

### Observation

The credit limit category simplifies customer segmentation based on available credit limits.

## Step 13.2 – PIN Change Status

### Purpose

This feature identifies whether the customer's card PIN has been changed after the account was opened.

```python
print(df["acct_open_date"].dtype)
```

```python
df["pin_change_status"] = np.where(
    df["year_pin_last_changed"] > df["acct_open_date"].dt.year,
    "Updated",
    "Not Updated"
)

df[["acct_open_date", "year_pin_last_changed", "pin_change_status"]].head()
```

## Step 13.3 – Multiple Cards Holder

### Purpose

Customers are classified based on the number of cards issued.

```python
df["multiple_cards_holder"] = np.where(
    df["num_cards_issued"] > 1,
    "Yes",
    "No"
)

df[["num_cards_issued","multiple_cards_holder"]].head()
```

### Purpose

The newly created features are reviewed to ensure they have been generated successfully.

```python
df[[
    "credit_limit_category",
    "pin_change_status",
    "multiple_cards_holder"
]].head()
```

## Step 14.1 – Calculate Key Performance Indicators (KPIs)

The following code calculates the primary business metrics used throughout the dashboard.

```python
total_customers = df["client_id"].nunique()

total_cards = len(df)

total_card_brands = df["card_brand"].nunique()

total_card_types = df["card_type"].nunique()

average_credit_limit = df["credit_limit"].mean()

maximum_credit_limit = df["credit_limit"].max()

chip_enabled_cards = (df["has_chip"] == "YES").sum()

non_chip_cards = (df["has_chip"] == "NO").sum()

cards_on_dark_web = (df["card_on_dark_web"] == "Yes").sum()
```

## Step 14.2 – Display KPI Summary

The calculated KPIs are displayed below to provide an overall summary of the dataset.

```python
print("="*50)
print("FINANCIAL TRANSACTIONS DASHBOARD KPI SUMMARY")
print("="*50)

print(f"Total Customers        : {total_customers}")
print(f"Total Cards            : {total_cards}")
print(f"Total Card Brands      : {total_card_brands}")
print(f"Total Card Types       : {total_card_types}")
print(f"Average Credit Limit   : ${average_credit_limit:,.2f}")
print(f"Maximum Credit Limit   : ${maximum_credit_limit:,.2f}")
print(f"Chip Enabled Cards     : {chip_enabled_cards}")
print(f"Non-Chip Cards         : {non_chip_cards}")
print(f"Cards on Dark Web      : {cards_on_dark_web}")
```

## Step 14.3 – KPI Summary Table

The following table presents all important business metrics in a structured format.

```python
kpi_summary = pd.DataFrame({
    "KPI": [
        "Total Customers",
        "Total Cards",
        "Total Card Brands",
        "Total Card Types",
        "Average Credit Limit",
        "Maximum Credit Limit",
        "Chip Enabled Cards",
        "Non-Chip Cards",
        "Cards on Dark Web"
    ],
    "Value": [
        total_customers,
        total_cards,
        total_card_brands,
        total_card_types,
        round(average_credit_limit,2),
        maximum_credit_limit,
        chip_enabled_cards,
        non_chip_cards,
        cards_on_dark_web
    ]
})

kpi_summary
```
